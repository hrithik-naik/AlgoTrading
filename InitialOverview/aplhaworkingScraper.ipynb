```python
#Fetching company with respect to its BSE CODE

from pandas import read_csv
company_name="Aris International Ltd."
data=read_csv("BSE_Companies.csv")
def get_bse_code(company_name):
    result = data[data['Company Name'] == company_name]['BSE Number']
    if not result.empty:
        return result.values[0]
    else:
        return "Company not found"
bsecodeofCurrentIteration=get_bse_code(company_name)
link="https://www.screener.in/company/"+bsecodeofCurrentIteration
```
