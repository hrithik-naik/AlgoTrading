```python
import pandas as pd

# The raw data as a string for demonstration (replace with your actual data)

data=test
# Split the data into lines and parse each line into components
lines = data.strip().split('\n')
parsed_data = []

for line in lines:
    parts = line.rsplit(' ', 2)  # Split from the right to handle multi-word company names
    company_name = parts[0]
    bse_number = parts[1]
    group = parts[2]
    parsed_data.append([company_name, bse_number, group])

# Create a DataFrame and save it to CSV
df= pd.DataFrame(parsed_data, columns=["Company Name", "BSE Number", "Group"])
df.to_csv("BSE_Companies.csv", index=False)

print("Data successfully converted to CSV!")
```
