```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

def get_wikipedia_summary(company_name):
    # Set up the WebDriver (change path to your ChromeDriver location)
    driver_path = '/path/to/chromedriver'  # Replace with the path to your WebDriver
    driver = webdriver.Chrome(driver_path)

    try:
        # Open Wikipedia
        driver.get('https://en.wikipedia.org')

        # Locate the search bar, enter the company name, and submit the search
        search_box = driver.find_element(By.NAME, 'search')
        search_box.send_keys(company_name)
        search_box.send_keys(Keys.RETURN)

        # Wait for the page to load
        time.sleep(2)

        # Find the summary paragraph (usually the first paragraph after the infobox)
        summary_paragraph = driver.find_element(By.CSS_SELECTOR, 'p')
        summary_text = summary_paragraph.text
        print(f"Summary for {company_name}:\n{summary_text}")

    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        # Close the browser
        driver.quit()

# Example usage
get_wikipedia_summary('Hyundai')
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import traceback

def get_full_wikipedia_content(company_name):
    driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
    driver.maximize_window()
    
    try:
        driver.get('https://en.wikipedia.org')

        # Wait for the search bar to load and enter the company name
        search_box = WebDriverWait(driver, 2).until(
            EC.element_to_be_clickable((By.NAME, 'search'))
        )
        search_box.send_keys(company_name)
        search_box.send_keys(Keys.RETURN)

        # Wait for the page to load
        WebDriverWait(driver,10).until(
            EC.presence_of_element_located((By.ID, 'firstHeading'))
        )

        # Retrieve all paragraphs in the main content section
        paragraphs = WebDriverWait(driver, 2).until(
            EC.visibility_of_all_elements_located((By.CSS_SELECTOR, 'div.mw-parser-output > p'))
        )

        # Concatenate the text from all paragraphs
        full_text = "\n".join(paragraph.text.strip() for paragraph in paragraphs if paragraph.text.strip())

        return full_text

    except Exception as e:
        print("An error occurred:")
        traceback.print_exc()
        return None

    finally:
        driver.quit()

# Example usage
content = get_full_wikipedia_content("Reliance Industries")
if content:
    print(content)
```
